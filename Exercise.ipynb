```python
def confusionMatrix(y,y_preds):
    TN = TP = FN = FP = 0
    for i in range(len(y)):
        if y[i][0] == 0 and int(y_preds[i][0]) == 0 :
            TN += 1 
        elif y[i][0] == 0 and int(y_preds[i][0]) == 1 :
            FP += 1
        elif y[i][0] == 1 and int(y_preds[i][0]) == 1 :
            TP += 1
        else :
            FN +=1
    return np.array([[TN,FP],[FN,TP]])
```

```python
def auc_of_roc(y,y_preds):
    TN = TP = FN = FP = 0
    for i in range(len(y)):
        if y[i][0] == 0 and int(y_preds[i][0]) == 0 :
            TN += 1 
        elif y[i][0] == 0 and int(y_preds[i][0]) == 1 :
            FP += 1
        elif y[i][0] == 1 and int(y_preds[i][0]) == 1 :
            TP += 1
        else :
            FN +=1
        
    tpr = TP / (TP + FN)
    fpr = 1 - (TN / (TN + FP))
    return (1 + tpr - fpr) / 2 
```

```python
def precisions_recalls(y,y_preds):
    precisions , recalls = [] , [] 
    y_preds_copy = y_preds.copy()
    for i in range(len(y_preds)):
        if y_preds[i][0] < .5 :
            y_preds[i][0] = 0
        else:
            y_preds[i][0] = 1
    precisions.append(precision(y,y_preds))
    recalls.append(recall(y,y_preds))
    for i in range(len(y_preds_copy)):
        if y_preds_copy[i][0] < .7:
            y_preds_copy[i][0] = 0
        else:
            y_preds_copy[i][0] = 1
    precisions.append(precision(y,y_preds_copy))
    recalls.append(recall(y,y_preds_copy))
    
    return precisions , recalls    
```
